##**Search algorithms with visited list**





###Part 1 : Breadth First Search Algorithm Function

*   input: Matrix list, start, goal
*   output: time complexity, space complexity, path, flag

```python
def breadthFirstSearch(graph,s,g):
  queue = [[s]]
  path = []
  visited = [s]
  flag = False
  tIME_calc = 0
  space_complex_calc = 0

  while queue:
    if space_complex_calc < len(queue):
      space_complex_calc = len(queue)
    sN = queue.pop(0)
    if sN[0] == g:
      path = sN
      queue = []
      flag = True
    else:
      sN1 = sN.copy()
      for i in sN:
          for j in graph[i]:
            if j not in visited:
              visited.append(j)
              sN1.insert(0,j)
              queue.append(sN1.copy())
              tIME_calc += 1
              sN1.pop(0)
      sN1 = []

  spc = max(len(visited),space_complex_calc)       
  return tIME_calc, spc, path, flag
```

###Part 2 : Depth First Search Algorithm Function

*   input: Matrix list, start, goal
*   output: time complexity, space complexity, path, flag

```python
def depthFirstSearch(graph,s,g):
  queue = [[s]]
  path = []
  visited = [s]
  flag = False
  tIME_calc = 0
  space_complex_calc = 0

  while queue:
    if space_complex_calc < len(queue):
      space_complex_calc = len(queue)
    sN = queue.pop(0)
    if sN[0] == g:
      path = sN
      queue = []
      flag = True
    else:
      sN1 = sN.copy()
      for j in graph[sN1[0]]:
        if j not in visited:
          visited.append(j)
          sN1.insert(0,j)
          queue.insert(0,sN1.copy())
          if queue[0][0] == g:
            path = sN1
            queue = []
            flag = True
            break
          tIME_calc += 1
          sN1.pop(0)

  spc = max(len(visited),space_complex_calc)
  return tIME_calc, spc, path, flag
```

###Part 3 : Best First Search Algorithm Function

*   input: Matrix list, cost, start, goal
*   output: time complexity, space complexity, path, flag

```python
def bestFirstSearch(graph, cost, s, g):

  pqueue = [[int(cost[s]),s]] # Priority Queue
  path = []
  visited = [s]
  flag = False
  tmp = -1
  tIME_calc = 0
  space_complex_calc = 0

  while pqueue:
    if space_complex_calc < len(pqueue):
      space_complex_calc = len(pqueue)
    sN = pqueue.pop(0)
    if sN[1] == g:
      path = sN[1:]
      pqueue = []
      flag = True
    else:
      sN.pop(0) #cost ro bardashtam
      sN1 = sN.copy()
      for j in sN:
          for i in graph[j]:
            if i not in visited:
              visited.append(i)
              sN1.insert(0,i)
              cost_i_value = int(cost[i])
              sN1.insert(0, cost_i_value)
              f = True
              for k in range(len(pqueue)):
                if cost_i_value <= int(pqueue[k][0]):
                  pqueue.insert(k, sN1.copy())
                  tIME_calc += 1
                  f = False
                  break
              if f:
                pqueue.append(sN1.copy())
                tIME_calc += 1
              sN1.pop(0)
              sN1.pop(0)
  
  spc = max(len(visited),space_complex_calc)
  return tIME_calc, spc, path, flag
```

###Part 4 : Progressive Deepening (Iterative Deepening) Search Algorithm Function


1.   depthFirstSearch_ddfs function:
  *   input: Matrix list, start, goal, depth
  *   output: time complexity, space complexity, path, flag

2.   iterativeDDFS function:
  *   input: Matrix list, start, goal, max depth
  *   output: time complexity, space complexity, path, flag

```python
def depthFirstSearch_ddfs(graph,s,g,depth):
  queue = [[s]]
  path = []
  visited = [s]
  flag = False
  tIME_calc = 0
  space_complex_calc = 0

  while queue:
    if space_complex_calc < len(queue):
      space_complex_calc = len(queue)
    
    sN = queue.pop(0)
    if sN[0] == g:
      path = sN
      queue = []
      flag = True

    if depth <= 0:
      flag = False
      spc = max(len(visited),space_complex_calc)
      return tIME_calc, spc, path, flag

    else:
      depth -= 1
      sN1 = sN.copy()
      if sN1[0] < len(graph):
        for j in graph[sN1[0]]:
          if j not in visited:
            visited.append(j)
            sN1.insert(0,j)
            queue.insert(0,sN1.copy())
            if queue[0][0] == g:
              path = sN1
              queue = []
              flag = True
              break
            tIME_calc += 1
            sN1.pop(0)
        sN1 = []

  spc = max(len(visited),space_complex_calc)
  return tIME_calc, spc, path, flag
```

```python
def iterativeDDFS(graph, source, goal, maxDepth):
  if maxDepth > len(graph)-1:
    print("Choose another depth")
    return '', False
  for i in range(maxDepth):
    tIME_calc = 0
    spc = 0
    tIME_calc1, spc1, path, flag = depthFirstSearch_ddfs(graph[:i+1], source, goal,i+1)
    tIME_calc += tIME_calc1
    spc += spc1
    if flag:
      return tIME_calc, spc, path, True
  return tIME_calc, spc, path, False
```

###Part 5.4 : Progressive Deepening (Iterative Deepening) Search Algorithm Function

**without visited list**

1.   wo_depthFirstSearch_ddfs function:
  *   input: Matrix list, start, goal, depth
  *   output: time complexity, space complexity, path, flag

2.   wo_iterativeDDFS function:

  *   output: time complexity, space complexity, path, flag

```python
def wo_depthFirstSearch_ddfs(graph,s,g,depth):
  queue = [[s]]
  path = []
  flag = False
  tIME_calc = 0
  space_complex_calc = 0

  while queue:
    if space_complex_calc < len(queue):
      space_complex_calc = len(queue)
    
    sN = queue.pop(0)
    if sN[0] == g:
      path = sN
      queue = []
      flag = True

    if depth <= 0:
      flag = False
      return tIME_calc, space_complex_calc, path, flag

    else:
      depth -= 1
      sN1 = sN.copy()
      if sN1[0] < len(graph):
        for j in graph[sN1[0]]:
            sN1.insert(0,j)
            queue.insert(0,sN1.copy())
            if queue[0][0] == g:
              path = sN1
              queue = []
              flag = True
              break
            tIME_calc += 1
            sN1.pop(0)
        sN1 = []

  return tIME_calc, space_complex_calc, path, flag
```

```python
def wo_iterativeDDFS(graph, source, goal, maxDepth):
  if maxDepth > len(graph)-1:
    print("Choose another depth")
    return '', False
  for i in range(maxDepth):
    tIME_calc = 0
    spc = 0
    tIME_calc1, spc1, path, flag = wo_depthFirstSearch_ddfs(graph[:i+1], source, goal,i+1)
    tIME_calc += tIME_calc1
    spc += spc1
    if flag:
      return tIME_calc, spc, path, True
  return tIME_calc, spc, path, False
```
